# Stag Hunt

![Stag Hunt](images/stag-hunt.png)

Stag Hunt represented as simultaneous and sequential game with prior reward.

Player2/Player1 choice depends on Player1/Player2 choice. In case players can reach agreement to play Stag/Hare strategy it wouldn't be profitable to switch strategy for neither player.

If Players agree to hunt Stag, both get reward 3. If Player1/Player2 switches from Stag to Hare strategy, he gets reward 2 and Player2/Player1 gets 0, thus the deviation from the agreement is not profitable for neither players. If Players cannot agree, a game could be represented as sequence with prior reward and player's choice could be assessed with forward induction. If Player1 picked reward, he reached terminal node and Player2 could play Hare strategy and get reward 2. If Player1 didn't pick a reward, Player2 can assume that Player1 picked a strategy with higher possible outcome, thus it is reasonable for Player2 to pick Stag strategy.
