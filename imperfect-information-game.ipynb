# Coin Toss Game

![coin toss](images/coin_toss.png)

The example game of Coin Toss. C represents a chance node. S is a Player2 (P2) subgame. The dotted line between the two P2 nodes means that P2 cannot distinguish between them (i.e. cannot know which P1's previous action led to observed subgame)

## Equations for solving Mixed Equilibria

### Players Strategies

\begin{align}
S_1 & = \{SellSell, SellPlay, PlaySell, PlayPlay\} \\
S_2 & = \{Heads, Tails\} \\
\end{align}

### Expected Utilities

#### Expected Utilities for (SellSell, Heads) and (SellSell, Tails) (Player2 action is irrelevant)

\begin{align}
E(P_1) & = 0.5 \cdot 0.5 + 0.5 \cdot (-0.5) = 0 \\
E(P_2) & = 0.5 \cdot (-0.5) + 0.5 \cdot 0.5 = 0 \\
\end{align}

#### Expected Utilities for (SellPlay, Heads)

\begin{align}
E(P_1) & = 0.5 \cdot 0.5 + 0.5 \cdot 1 = 0.75 \\
E(P_2) & = 0.5 \cdot (-0.5) + 0.5 \cdot (-1) = -0.75 \\
\end{align}

#### Expected Utilities for (SellPlay, Heads)

\begin{align}
E(P_1) & = 0.5 \cdot 0.5 + 0.5 \cdot 1 = 0.75 \\
E(P_2) & = 0.5 \cdot (-0.5) + 0.5 \cdot (-1) = -0.75 \\
\end{align}

#### Expected Utilities for (SellPlay, Tails)

\begin{align}
E(P_1) & = 0.5 \cdot 0.5 + 0.5 \cdot (-1) = -0.25 \\
E(P_2) & = 0.5 \cdot (-0.5) + 0.5 \cdot 1 = 0.25 \\
\end{align}

#### Expected Utilities for (PlayPlay, Heads)

\begin{align}
E(P_1) & = 0.5 \cdot (-1) + 0.5 \cdot 1 = 0 \\
E(P_2) & = 0.5 \cdot 1 + 0.5 \cdot (-1) = 0 \\
\end{align}

#### Expected Utilities for (PlayPlay, Tails)

\begin{align}
E(P_1) & = 0.5 \cdot 1 + 0.5 \cdot (-1) = 0 \\
E(P_2) & = 0.5 \cdot (-1) + 0.5 \cdot 1 = 0 \\
\end{align}

### Payoff Matrix
\begin{bmatrix}
    (0, 0)         & (0, 0) \\
    (0.75, -0-75)  & (-0.25, 0.25) \\
    (-0.75, 0.75)  & (0.25, -0.25) \\
    (0, 0)         & (0, 0)
\end{bmatrix}

Rows 1 and 4 can be eliminated to get the following Payoff Matrix:

\begin{bmatrix}
    (0.75, -0-75)  & (-0.25, 0.25) \\
    (-0.75, 0.75)  & (0.25, -0.25)
\end{bmatrix}

Row 1 contains expected values for Player1 SellPlay strategy

Row 2 contains expected values for Player1 PlaySell strategy

Column 1 contains expected values for Player2 Heads strategy

Column 2 contains expected values for Player2 Tails strategy

### Calculating Optimal Strategy for Player2 subgame S

Were P2 to always guess Heads, P1 would receive \$0.5 for choosing Sell when coin lands Heads, and \$1 for Play when it lands Tails. This would result average \$0.75 for P1 and \$0.25 for P2.

P2 has to find the random guessing ratio for guessing Heads/Tails such that the Expected value of Sell for P1 is equal to Play (P1 should not be able to take advantage of any action).

\begin{align}
E(Sell) &= E(Play) \\
E(Sell) &= 0.75 \cdot \rho_{Heads} - 0.25 \cdot (1 - \rho_{Heads}) \\
E(Play) &= -0.75 \cdot \rho_{Heads} + 0.25 \cdot (1 - \rho_{Heads}) \\
\end{align}

\begin{align}
0.75 \cdot \rho_{Heads} - 0.25 \cdot (1 - \rho_{Heads}) &= -0.75 \cdot \rho_{Heads} + 0.25 \cdot (1 - \rho_{Heads}) \\
\rho_{Heads} &= 0.25
\end{align}

If P2 Guessed Heads with 25% probability and Tails with 75% probability, P1 could only receive \$0.5 on average instead of $0.75
